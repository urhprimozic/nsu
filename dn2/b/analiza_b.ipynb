```python
import pandas as pd
import arff
import matplotlib.pyplot as plt 
from pandas.api.types import is_numeric_dtype

df = pd.read_csv('dn2b.csv')
for f in df.columns:
    if f in ['f1', 'f2', 'f3']:
        df[f].plot()
        plt.title(f'Feature {f}')
        plt.show()
    else:
        df[f].hist()
        plt.title(f'Feature {f}')
        plt.show()

```
